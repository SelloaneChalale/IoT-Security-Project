```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import os

# List of attack types and benign label
attack_types = ["Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", "DoS slowloris", "FTP-Patator",
                "Heartbleed", "Infiltration", "PortScan", "SSH-Patator", "Web Attack - Brute Force",
                "Web Attack - Sql Injection", "Web Attack - XSS"]
benign_type = "BENIGN"

# Feature selection function
def perform_feature_selection(data, label_column="Label"):
    # Ensure consistent column naming
    data.columns = data.columns.str.strip()  # Remove any leading/trailing spaces
    if label_column not in data.columns:
        raise KeyError(f"'{label_column}' column not found in dataset")

    # Separate features and target variable
    X = data.drop(columns=[label_column])
    y = data[label_column].apply(lambda x: 1 if x != benign_type else 0)
    
    # Fit RandomForestClassifier for feature importance
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X, y)
    
    # Extract and return feature importances and feature names
    importances = clf.feature_importances_
    return importances, X.columns

# Iterate over attack types and process each file
for attack_type in attack_types:
    input_filename = f"{attack_type}_vs_{benign_type}.csv"
    
    # Load and process the dataset
    try:
        attack_data = pd.read_csv(input_filename, low_memory=False)
        
        # Perform feature selection and obtain importances
        importances, features = perform_feature_selection(attack_data)
        
        # Create a DataFrame for importance results
        importance_df = pd.DataFrame({"Feature": features, "Importance": importances})
        total_importance = importance_df["Importance"].sum()
        importance_df["Percentage"] = importance_df["Importance"] / total_importance * 100
        importance_df = importance_df.sort_values(by="Importance", ascending=False)
        
        # Print and save the top 20 features
        print(f"\nTop 20 features and their percentages for {attack_type}:")
        print(importance_df.head(20))
        
        # Save importance data to a CSV file
        importance_filename = f"{attack_type}_importance.csv"
        importance_df.to_csv(importance_filename, index=False)
        print(f"Saved importance list for {attack_type}")
        
        # Create a bar plot for top 20 features' importance
        plt.figure(figsize=(10, 6))
        top_20_df = importance_df.head(20)
        top_20_df.plot(kind="bar", x="Feature", y="Importance", legend=None)
        plt.title(f"Feature Importance for {attack_type}")
        plt.xlabel("Feature")
        plt.ylabel("Importance")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
        
    except FileNotFoundError:
        print(f"File {input_filename} not found.")
    except KeyError as e:
        print(e)

print("Feature selection and visualization completed for all attack types.")

# Process combined data for overall importance
combined_data = pd.read_csv("combined_data.csv")
importance_list, features = perform_feature_selection(combined_data, label_column=" Label")

# Create DataFrame for combined data's feature importance
importance_df = pd.DataFrame({"Feature": features, "Importance": importance_list})
total_importance = importance_df["Importance"].sum()
importance_df["Percentage"] = importance_df["Importance"] / total_importance * 100
importance_df = importance_df.sort_values(by="Importance", ascending=False)

# Print and save top 20 features
print("\nTop 20 features and their percentages for combined_data:")
print(importance_df.head(20))

# Save importance list to CSV
importance_filename = "combined_data_importance.csv"
importance_df.to_csv(importance_filename, index=False)
print("Saved importance list for combined_data")

# Plot feature importance for combined data
plt.figure(figsize=(10, 6))
top_20_df = importance_df.head(20)
top_20_df.plot(kind="bar", x="Feature", y="Importance", legend=None)
plt.title("Feature Importance for combined_data")
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#adapted from https://github.com/yasakrami/Threat-Detection-in-Cyber-Security-Using-AI
```
