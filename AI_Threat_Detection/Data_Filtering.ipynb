```python
import pandas as pd
import time

# Record start time
start_time = time.time()

# Load the main dataset (adjust path if necessary)
main_dataset = pd.read_csv("C:/Users/PC/combined_data.csv")

# Define attack types and BENIGN
attack_types = ["Bot", "DDoS", "DoS GoldenEye", "DoS Hulk", "DoS Slowhttptest", 
                "DoS slowloris", "FTP-Patator", "Heartbleed", "Infiltration", 
                "PortScan", "SSH-Patator", "Web Attack  Brute Force", 
                "Web Attack  Sql Injection", "Web Attack  XSS"]
benign_type = "BENIGN"

def save_attack_vs_benign(attack_type, benign_type, dataset):
    attack_data = dataset[dataset[" Label"] == attack_type]
    benign_data = dataset[dataset[" Label"] == benign_type]
    combined_data = pd.concat([attack_data, benign_data], axis=0).sample(frac=1, random_state=42)
    output_filename = f"{attack_type}_vs_{benign_type}.csv"
    combined_data.to_csv(output_filename, index=False)
    print(f"Saved {output_filename}")

# Generate files for each attack type
for attack_type in attack_types:
    save_attack_vs_benign(attack_type, benign_type, main_dataset)

# Calculate and print the execution time
end_time = time.time()
print(f"Execution time: {end_time - start_time:.2f} seconds")

# File Analysis
def analyze_files(file_names):
    for file_name in file_names:
        data = pd.read_csv(file_name)
        num_benign = (data[' Label'] == 'BENIGN').sum()
        num_attack = (data[' Label'] != 'BENIGN').sum()
        print(f"File: {file_name}")
        print(f"Benign instances: {num_benign}, Attack instances: {num_attack}")
        print(f"Shape: {data.shape}\n{'-'*30}")

# List of generated filenames
generated_files = [f"{attack}_vs_{benign_type}.csv" for attack in attack_types]

# Analyze generated files
analyze_files(generated_files)

#adapted from https://github.com/yasakrami/Threat-Detection-in-Cyber-Security-Using-AI
```
